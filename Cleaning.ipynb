```python
import pandas as pd
```

```python
spotify = pd.read_csv('spotify.csv', sep=',')
spotify
```

```python
artists_expanded = spotify['artists'].str.split(r'\s*,\s*', expand=True)
```

```python
artists_expanded.columns = [f'Artist_{i+1}' for i in range(artists_expanded.shape[1])]
```

```python
spotify = pd.concat([spotify, artists_expanded], axis=1)
```

```python
spotify
```

```python
spotify['album'] = spotify['album'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
```
